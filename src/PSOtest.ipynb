```python
bestInd = Batch.loadBestInd("expPSO/pso_DOG_competitionbestInd.pi")
```

```python
def loadParams(paramList):    
    parameters = {}
    for param in paramList:
        parameters[param] = [d[param] for d in bestInd]
    return parameters



```

```python
params = loadParams(['iExc','iInh','wExc','wInh','h','th','tau'])
```
